```python
def get_season_squad_data(season):
  excel_file = pd.ExcelFile('/content/gdrive/MyDrive/APS360_Project/Data/{}.xlsx'.format(season))
  squad_sheets = [
    'Squad Data stats',
 'Squad Data keepers',
 'Squad Data shooting',
 'Squad Data defense',
 'Squad Data possession']

  dataframes = {}
  squad_data = pd.DataFrame()
  for sheet_name in squad_sheets:
      dataframes[sheet_name] = pd.read_excel(excel_file, sheet_name, header = 1)
      dataframes[sheet_name].drop(0, inplace = True)
      dataframes[sheet_name].drop(columns=['Unnamed: 0'], inplace=True)
      assert(dataframes[sheet_names[0]].shape[0]==dataframes[sheet_name].shape[0])
      if sheet_name == squad_sheets[0]:
        squad_data = dataframes[sheet_names[0]]
      elif sheet_name == 'Squad Data keepers':
         squad_data = pd.merge(squad_data, dataframes[sheet_name], how = 'inner', on=['90s', 'Squad'], suffixes=[None, '_{}'.format(sheet_name)])
      else:
         squad_data = pd.merge(squad_data, dataframes[sheet_name], how = 'inner', on=['# Pl', '90s', 'Squad'], suffixes=[None, '_{}'.format(sheet_name)])
  return squad_data
```

```python
seasons_dataframes = {}
for season in seasons:
  seasons_dataframes[season] = get_season_squad_data(season)
  seasons_dataframes[season].drop(columns=['Starts','Min','90s',  'G+A',
 'G-PK', 'xG',
 'npxG',
 'xAG',
 'npxG+xAG', 'Gls.1',
 'Ast.1',
 'G+A.1',
 'G-PK.1',
 'G+A-PK',
 'xG.1',
 'xAG.1',
 'xG+xAG',
 'npxG.1',
 'npxG+xAG.1', 'Save%',
'GA90',
'CS%', 'Gls_Squad Data shooting', 'SoT%',
 'Sh/90',
 'SoT/90',
 'G/Sh',
 'G/SoT','PKatt_Squad Data keepers','PK_Squad Data shooting','xG_Squad Data shooting',
 'npxG_Squad Data shooting',
 'npxG/Sh',
 'G-xG',
 'np:G-xG','Tkl%','Succ%',
 'Tkld%', 'PrgC_Squad Data possession'

], inplace=True)
```

```python
ranges = [(n, n+1) for n in range(2017, 2023)]
seasons = ["{}-{}".format(year1,year2) for (year1, year2) in ranges]
seasons_matchdata = {}

def get_data_for_season(season):
    with pd.ExcelFile(f'/content/gdrive/MyDrive/APS360_Project/Data/{season}.xlsx') as excel_file:
        fixtures_df = pd.read_excel(excel_file, 'Fixture data')
        fixtures_df.drop(columns=['Unnamed: 0', 'Day', 'Date', 'Time','Venue','Referee', 'Match Report', 'Notes', 'Attendance'], inplace=True)
        fixtures_df = fixtures_df.dropna(how='all')
        fixtures_df['winner'] = fixtures_df['Score'].apply(compare_scores)
        fixtures_df.drop(columns=['Score'], inplace=True)
        merged_df = pd.merge(fixtures_df, seasons_dataframes[season], left_on='Home', right_on='Squad', how='left')
        merged_df = pd.merge(merged_df, seasons_dataframes[season], left_on='Away', right_on='Squad', suffixes=('_Home', '_Away'), how='left')
        merged_df.drop(columns=['Home', 'Away'], inplace=True)
        seasons_matchdata[season] = merged_df

for season in seasons:
    get_data_for_season(season)
```

```python

excel_file = '/content/gdrive/MyDrive/APS360_Project/Data/match_data_refined.xls'
with pd.ExcelWriter(excel_file) as writer:
    # Iterate over each key-value pair in the dictionary
    for sheet_name, df in seasons_matchdata.items():
        # Write each DataFrame to a separate sheet
        df.to_excel(writer, sheet_name=sheet_name, index=False)
```
